## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
# URL of the page to scrape
url = "https://en.wikipedia.org/wiki/Jakub_Maxa"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = bs4.BeautifulSoup(response.content, "html.parser")

# Find all the text content within the HTML tags
text_content = soup.get_text()

# Print the text content
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
# Extract the image URL from the image tag
image_url = "https://cdn.pixabay.com/photo/2015/04/23/22/00/tree-736885_960_720.jpg"

# Send a GET request to download the image
image_response = requests.get(image_url)

# Save the image to a file in the current directory
with open('image.png', 'wb') as file:
    file.write(image_response.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
# Find all the links in the page
links = soup.find_all('a')

# Print the first 100 characters of ten links
for link in links[:10]:
    print(link.get('href')[:100])
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
# Function to download the first 1000 characters of a script given a link
def download_script(link):
    # Send a GET request to the URL
    try:
        response = requests.get(link)
    except:
        return None
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = bs4.BeautifulSoup(response.content, "html.parser")
    
    # Find all the text content within the HTML tags
    text_content = soup.get_text()
    
    # Return the first 1000 characters of the text content
    return text_content[:1000]

# Download the first 1000 characters of each script
characters = None
characters = scripts['link'].apply(download_script)
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
df = characters
df.to_csv('pudding_texts.csv', index=False)
```
